```python
from google.colab import drive
drive.mount('/content/drive')
os.chdir('/content/drive/My Drive/Badanie2')
```

```python
custom_transforms = transforms.Compose([
transforms.ToPILImage(),
transforms.Resize(900),
transforms.RandomHorizontalFlip(0.5),
transforms.ColorJitter(saturation=0.2),
transforms.ToTensor(),
utils.normalize_transform(),
])


```

```python
Train_dataset=core.Dataset('/content/drive/My Drive/Badanie2/Train', transform=custom_transforms)
Test_dataset = core.Dataset('/content/drive/My Drive/Badanie2/Test')
# ścieżki do folderów ze zdjęciami treningowymi i testowymi
loader=core.DataLoader(Train_dataset, batch_size=2, shuffle=True)
model = core.Model(["wilk"]) # określamy ile mamy nazw obiektów oraz wpisujemy ich nazwy
losses = model.fit(loader, Test_dataset, epochs=25, lr_step_size=5, learning_rate=0.001, verbose=True)
```

```python
plt.plot(losses)
plt.show()
```

```python
image = utils.read_image("/content/drive/My Drive/Badanie2/test.jpg") 
predictions = model.predict(image)
labels, boxes, scores = predictions
show_labeled_image(image, boxes, labels)
```

```python
thresh=0.3
filtered_indices=np.where(scores>thresh)
filtered_scores=scores[filtered_indices]
filtered_boxes=boxes[filtered_indices]
num_list = filtered_indices[0].tolist()
filtered_labels = [labels[i] for i in num_list]
show_labeled_image(image, filtered_boxes, filtered_labels)
```
